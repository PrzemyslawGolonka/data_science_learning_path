```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
# Functional method of creating plots 
```

```python
plt.plot(x,y,'r-')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.show() # if we dont have jupiter notebook; it prints the plot instead of showing it
```

```python
plt.subplot(1,2,1) #num of rows, num of columns, the plot we refering to
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object oriented methof, we create object and call method on in
```

```python
fig = plt.figure() # it creates a 'canvas'

axes = fig.add_axes([0.1,0.1,0.8,0.8]) #left, bottom, wight hight of exies

axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 =fig.add_axes([0.2,0.5,0.4,0.3]) #all must be between 0 and 1
#0.2 from left, if it would be 0.8 it would be 80% away from left side, so more on the right side 
axes1.plot(x,y)
axes2.plot(y,x)
axes2.set_title('Smaller plot')
axes1.set_title('Larger plot')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2) #axes object is just a list of mat plot lib axies 

for curr_ax in axes:
    curr_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```

```python
fig = plt.figure(figsize=(8,2)) #tuple of width and high
#dpi dorts per inch, deafult 100

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig('my_pic.png') # how to save a figure, we just add the extenson of the file f. e .png,.pdf
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1]) # localtion of axies

ax.plot(x,x**2, label='X Squered') # we have to specify labels before creating a legend
ax.plot(x,x**3, label='X Cubed')

ax.legend(loc=0) #at the bottom of the code; loc - where the legend will be displayed or tuple with parameters of x and y axies
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',linewidth=3, linestyle='--',marker='o',markersize=20, markerfacecolor='yellow', markeredgewidth=3,markeredgecolor='green') #rgb hex code
# we can put lw and ls for linewidth and line style
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',linewidth=3,linestyle='--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
