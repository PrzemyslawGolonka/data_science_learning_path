```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
df = pd.read_csv('../DATA/fake_reg.csv')
```

```python
df.head()
```

```python
sns.pairplot(df)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['feature1','feature2']].values #tf takes np arrays not pd data frames thats why we takie .values
```

```python
y = df['price'].values 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
#next we have to normalize our data 
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train) #only x train!
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
#help(Sequential)
```

```python
#how to build keras model
#model = Sequential([Dense(4,activation='relu'),
                   #Dense(2,activation='relu'),
                   #Dense(1)]) #inot n of neurons, activaton - which activation functin should we use 
```

```python
#this is preferred, easier to change or delete some parts 
model = Sequential()

model.add(Dense(4,activation='relu'))
model.add(Dense(4,activation='relu'))
model.add(Dense(4,activation='relu'))

model.add(Dense(1)) #one price output 

model.compile(optimizer='rmsprop',loss='mse') #optimizer adam, gradient descent ect; loss - how to choose right one is down on the notebook
```

```python
model.fit(x=X_train,y=y_train,epochs=250)
```

```python
loss_df = pd.DataFrame(model.history.history)
```

```python
loss_df.plot()
```

```python
model.evaluate(X_test,y_test,verbose=0)
```

```python
model.evaluate(X_train,y_train,verbose=0)
```

```python
test_predictions = model.predict(X_test)
```

```python
test_predictions = pd.Series(test_predictions.reshape(300,))
```

```python
pred_df = pd.DataFrame(y_test,columns=['Test True Y'])
```

```python
pred_df = pd.concat([pred_df,test_predictions],axis=1)
```

```python
pred_df.columns = ['Test True Y', 'Model Predictions']
```

```python
sns.scatterplot(x='Test True Y',y='Model Predictions',data=pred_df)
#almost stright line means that out model does pretty well 
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
mean_absolute_error(pred_df['Test True Y'],pred_df['Model Predictions'])
# we are on avarage 4$ off on the price, which is around 1% of the price so its okay, it it was 50% oof the price then is bad 
```

```python
df.describe() #4$ is 1% of mean 
```

```python
mean_squared_error(pred_df['Test True Y'],pred_df['Model Predictions'])
```

```python
mean_squared_error(pred_df['Test True Y'],pred_df['Model Predictions'])**0.5
#root mean squered error
```
